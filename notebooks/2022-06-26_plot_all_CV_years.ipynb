```python
fig, ax = plt.subplots(5, 5, figsize=(20, 20), sharex=True, sharey=True)
ax = ax.flatten()

for _ax, year in zip(ax, range(1993, 2018)):
    df = subset_trajectories(
        load_cape_verde_trajectories(year=year),
        num_traj=1_000,
        use_random=True,
        random_seed=456,
    )
    df.plot.scatter(x="longitude", y="latitude", s=3, alpha=0.1, ax=_ax);
    _ax.set_title(f"{year:04d}")
    
```
